```python
##install ete3-
#conda install conda-forge::ete3
##load ete3-
import ete3
from ete3 import Tree
```

```python
#Read the newick file-http://etetoolkit.org/docs/latest/tutorial/tutorial_trees.html#reading-newick-trees
t = Tree("Extended_Data_S1/GH13_23.newick")
print(t.write())
```

```python
#Write tree to file
t.render("Extended_Data_S1/GH13_23.newick.pdf", w=183, units="mm")
```
